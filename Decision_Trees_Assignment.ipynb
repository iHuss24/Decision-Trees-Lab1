# Q3: Split tha dataset into training and testing

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
x = df.drop(['real_fake'], axis=1)
#target
y = df['real_fake']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30)
```

# Q4: Build three machine models 

## Q4.1: The first machine model
- Print the model's name.
- Print the model's accuracy.
- Print the model's confusion matrix.

```python
dtm = DecisionTreeClassifier()
print(dtm)
```

```python
dtm.fit(x_train,y_train)
dtyp = dtm.predict(x_test)
```

```python
print("Train data accuracy:",accuracy_score(y_true = y_train, y_pred=dtm.predict(x_train)))
print("Test data accuracy:",accuracy_score(y_true = y_test, y_pred=dtyp))
```

```python
print("Confusion Matrix for traning data")
print(confusion_matrix(y_train,dtm.predict(x_train)))
print("Confusion Matrix for test data")
print(confusion_matrix(y_test,dtyp))
```

## Q4.2: The second machine model
- Print the model's name.
- Print the model's accuracy.
- Print the model's confusion matrix.

```python
rfm = RandomForestClassifier()
print(rfm)
```

```python
rfm.fit(x_train, y_train);
yp= rfm.predict(x_test)
```

```python
print("Train data accuracy:",accuracy_score(y_true = y_train, y_pred=rfm.predict(x_train)))
print("Test data accuracy:",accuracy_score(y_true = y_test, y_pred=yp))
```

```python
print("Confusion Matrix for traning data")
print(confusion_matrix(y_train,rfm.predict(x_train)))
print("Confusion Matrix for test data")
print(confusion_matrix(y_test,yp))
```
